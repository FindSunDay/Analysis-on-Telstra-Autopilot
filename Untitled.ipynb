```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

follow = pd.read_csv('autopilot.csv')
follow
```

```python
condition1 = follow['Is_Front_Car_Change_Lane']==1
figure3 = follow[condition1]
fun1 = {'Is_Front_Car_Change_Lane':{'Total_Change': 'sum'}}
figure3 = figure3.groupby('Day_Or_Night').agg(fun1)
figure3 = figure3.reset_index()
figure3.columns = figure3.columns.droplevel(0)
figure3.rename(columns = {'':'Day_Or_Night'}, inplace = True)
figure3

ax = figure3.plot.bar(x = 'Day_Or_Night',y = 'Total_Change', rot = 0, label = 'Changes')
plt.xlabel('Time_Day_Or_Night')
plt.ylabel('Changes')
plt.show()
```

```python
condition2 = follow['Is_Front_Car_Change_Lane']==1
figure4 = follow[condition1]
figure4 = pd.melt(figure4, value_vars = ['Windy','Rainy'], var_name = 'Weather')
fun2 = {'Weather':{'Total_Change': 'count'}}
figure4 = figure4.groupby('Weather').agg(fun2)
figure4 = figure4.reset_index()
figure4.columns = figure4.columns.droplevel(0)
figure4.rename(columns = {'':'Weather'}, inplace = True)
figure4

ax = figure4.plot.bar(x = 'Weather',y = 'Total_Change', rot = 0, label = 'Changes')
plt.xlabel('Weather')
plt.ylabel('Changes')
plt.show()
```
